```python
df1["Country"] = df1["Country"].astype("category")
df1["Character"] = df1["Character"].astype("category")
df1["Character Gender"] = df1["Character Gender"].astype("category")
df1["Team"] = df1["Team"].astype("category")
df1.get_dtype_counts()
```

```python
country_lifetime_score = pd.DataFrame(df1["Country"].value_counts())
country_lifetime_score.columns = ["Lifetime Score"]
country_lifetime_score.plot(kind = "pie", y = "Lifetime Score", figsize=(10,40))
```
